```python
data='''name,gender,grade
john,m,76
jane,f,82
joseph,m,84
susan,f,72
'''

lines=data.strip().split('\n')
header=lines[0].split(',')
students=[]

for line in lines[1:]:
    name, gender, grade = line.split(',')
    students.append({
        'name': name,
        'gender': gender,
        'grade': int(grade)
    })

# Sort by grade in descending order
sorted_students = sorted(students, key=lambda x: x['grade'], reverse=True)

print("Students sorted by grade (High to Low):")
for student in sorted_students:
    print(f"Name: {student['name']}, Gender: {student['gender']}, Grade: {student['grade']}")
```

```python
numbers = []

while True:
    user_input = input("Enter a number (or '=' to stop): ")
    if user_input == '=':
        print('Stop Value')
        break
    try:
        number = float(user_input)
        numbers.append(number)
        print(f"Added: {number}. Current list: {numbers}")
    except ValueError:
        print("Invalid input. Please enter a valid number or '='.")
```

```python
if numbers:
  total=sum(numbers)
  print(f"The sum of the entered values is: {total}")
else:
    print("No numbers were entered to sum.")
```

### List Operations and Average Calculation

```python

myList=[10, 20, 30]
myList.append(40)
myList.insert(1, 15)
print("Modified list:", myList)
```

```python

total=sum(myList)
numItems = len(myList)

if numItems > 0:
    average = total/numItems
    print("Average score:", average)
else:
    print("The list is empty!!!")
```

A program taking any values till an equal sign is placed so that it stops to enter the value and print out the sum of the values, lowest of the value,highest of the value,average and numerical

```python
numbers = []

while True:
    user_input = input("Enter a number (or '=' to stop): ")
    if user_input == '=':
        break
    try:
        number = float(user_input)
        numbers.append(number)
    except ValueError:
        print("Invalid input. Please enter a valid number or '='.")

if numbers:
    total_sum = sum(numbers)
    lowest_value = min(numbers)
    highest_value = max(numbers)
    count = len(numbers)
    average = total_sum / count

    print(f"\nSum of values: {total_sum}")
    print(f"Lowest value: {lowest_value}")
    print(f"Highest value: {highest_value}")
    print(f"Average of values: {average}")
    print(f"Number of numerical values: {count}")
else:
    print("No numbers were entered.")
```
