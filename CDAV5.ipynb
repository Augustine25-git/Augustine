Negatives, Zeros and Positives

```python
numbers = []

while True:
    user = input("Enter a number (or '' to stop): ")
    if user == '':
        break
    try:
        number = float(user)
        numbers.append(number)
    except ValueError:
        print("Invalid input. Please enter a valid number or blank.")
```

```python
negatives = []
zeros = []
positives = []

for num in numbers:
    if num < 0:
        negatives.append(num)
    elif num == 0:
        zeros.append(num)
    else:
        positives.append(num)

print("Categorized Numbers:")
for num in negatives:
    print(int(num))
for num in zeros:
    print(int(num))
for num in positives:
    print(int(num))
```

```python
# Main program to manage student records (without functions or classes)
student_records = {}

while True:
    print("\n--- Student Record Program ---")
    print("1. Add New Student")
    print("2. Record Marks")
    print("3. View Failing Scores")
    print("4. View Complete Details")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ").strip()

    if choice == '1':
        # Add New Student
        name = input("Enter student name: ").strip()
        student_id = input("Enter student ID: ").strip()
        if name and student_id:
            if student_id not in student_records:
                student_records[student_id] = {'name': name, 'marks': {}}
                print(f"Student {name} with ID {student_id} added.")
            else:
                print("Student ID already exists.")
        else:
            print("Name and Student ID cannot be empty.")

    elif choice == '2':
        # Record Marks
        student_id = input("Enter student ID: ").strip()
        if student_id in student_records:
            subject = input("Enter subject name: ").strip()
            try:
                score = float(input("Enter score for the subject: "))
                if 0 <= score <= 100:
                    student_records[student_id]['marks'][subject] = score
                    print(f"Mark for {student_records[student_id]['name']} in {subject} recorded as {score}.")
                else:
                    print("Invalid score. Score must be between 0 and 100.")
            except ValueError:
                print("Invalid score. Please enter a numeric value.")
        else:
            print("Student not found.")

    elif choice == '3':
        # View Failing Scores
        student_id = input("Enter student ID: ").strip()
        if student_id in student_records:
            student_info = student_records[student_id]
            failing_found = False
            print(f"\nFailing scores for {student_info['name']} (ID: {student_id}):")
            for subject, score in student_info['marks'].items():
                if score < 40: # Default passing score
                    print(f"  {subject}: {score}")
                    failing_found = True
            if not failing_found:
                print(f"No failing scores for {student_info['name']}.")
        else:
            print("Student not found.")

    elif choice == '4':
        # View Complete Details
        student_id = input("Enter student ID: ").strip()
        if student_id in student_records:
            student_info = student_records[student_id]
            details = f"\nStudent Name: {student_info['name']}\nStudent ID: {student_id}\nMarks:"
            if not student_info['marks']:
                details += " No marks recorded yet."
            else:
                for subject, score in student_info['marks'].items():
                    details += f"\n  {subject}: {score}"
            print(details)
        else:
            print("Student not found.")

    elif choice == '5':
        # Exit
        print("Exiting program. Goodbye!")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```
