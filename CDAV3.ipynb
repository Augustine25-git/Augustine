Caeser caipher

```python

message = input("Enter your message: ")
shift = int(input("Enter the shift amount (positive to encode, negative to decode): "))

shifted_message = ""
for char in message:
    if char.isupper():
        new_char = chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
        shifted_message += new_char
    elif char.islower():
        new_char = chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
        shifted_message += new_char
    else:
        shifted_message += char
print("Result:", shifted_message)
```

 Is a String a Palindrome?

```python

user= input("Enter a word to check: ")

cleaned_word = user.lower()
left = 0
right = len(cleaned_word) - 1
is_palindrome = True
while left < right:

    if cleaned_word[left] != cleaned_word[right]:
        is_palindrome = False
        break
    left += 1
    right -= 1
if is_palindrome:
    print(f'Success! "{user}" reads perfectly identical forward and backward.')
else:
    print(f'Notice: "{user}" is not a palindrome.')
```

Greatest Common Divisor

```python

num1 = int(input("Enter the first positive number: "))
num2 = int(input("Enter the second positive number: "))
if num1 < num2:
    d = num1
else:
    d = num2
while (num1 % d != 0) or (num2 % d != 0):
    d = d - 1
print("The Greatest Common Divisor is:", d)
```
